```python
img3=cv2.imread('crossword.jpg',0)
plt.imshow(img3,cmap='gray')
```

```python
def magnify(img):
  fig=plt.figure(figsize=(15,15))
  ax=fig.add_subplot(111)
  ax.imshow(img,cmap='gray')
```

```python
magnify(img3)
```

```python
ret2,thr2=cv2.threshold(img3,127,255,cv2.THRESH_BINARY)
```

```python
magnify(thr2)
```

```python
thr3=cv2.adaptiveThreshold(img3,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,8)
magnify(thr3)
```
